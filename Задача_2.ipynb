````python
import matplotlib.pyplot as plt
data_view = df['waterfront'].value_counts()
val = data_view.values
labels = data_view.index
plt.figure(figsize=(8, 8))
plt.pie(val, labels=['С видом на набережную', 'Без вида на набережную'],
        explode=[0, 0.3], radius=1.5, startangle=60,
        labeldistance=1.2, autopct='%.1f%%', colors=['#66b3ff','#99ff99'])

plt.title('Распределение домов по наличию вида на набережную')
plt.show()
````

Вывод:

*   Большинство недвижимости не имеет вида на набережную, что указывает на преобладание таких объектов на рынке.
*   Менее 30% домов имеют вид на набережную, что может свидетельствовать о высокой ценности таких объектов.Только 0,8% всей недвижимости продается с видом на набережную

````python
df.floors.max(), df.floors.min()
````

````python
plt.figure(figsize=(10, 6))
sns.histplot(df['floors'], bins=10, kde=True, color='skyblue')
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['floors'], color='lightgreen')
plt.title('Boxplot распределения этажей домов')
plt.xlabel('Количество этажей')
plt.grid(True)
plt.show()
````



```
Изучите распределение состояния домов
```



````python

condition_counts = df['condition'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(hue=condition_counts.index, y=condition_counts.values, palette='viridis')
plt.title('Распределение состояния домов')
plt.xlabel('Состояние домов (от 1 до 5)')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
````

Вывод:

*   Основное распределение этажей домов находится в диапазоне от 1 до 3 этажей, что характерно для большинства одно- и двухэтажных строений.
*   Многоэтажные дома (более 3 этажей) встречаются реже, что может свидетельствовать о том, что такие дома являются менее типичными для данного региона.
*   Boxplot показывает наличие небольшого числа домов с экстраординарно высоким количеством этажей, что говорит о возможных выбросах или особых архитектурных решениях.
*   Небольшое количество домов находится в плохом или очень плохом состоянии (оценки 1 и 2), что может свидетельствовать о необходимости ремонта или реконструкции.
