````python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')

plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
````

Вывод:

*   На гистограмме видно, что основная часть домов продается по цене до 1 миллиона долларов, с резким снижением количества объектов по мере увеличения стоимости. Это говорит о том, что премиальная недвижимость составляет меньшую часть рынка, в то время как большинство покупателей ориентируются на более доступные варианты.




````python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sqft_living'], y=df['price'], alpha=0.5)

plt.title('Зависимость стоимости от жилой площади')
plt.xlabel('Жилая площадь (кв. футы)')
plt.ylabel('Стоимость')
plt.grid(True)
plt.show()
````

Вывод:

*   На этом графике видно, что существует положительная корреляция между жилой площадью и стоимостью недвижимости: чем больше жилая площадь, тем дороже дом. Однако при значениях жилой площади более 5000 кв. футов наблюдается разброс цен, что может указывать на дополнительные факторы, влияющие на стоимость.


````python
plt.figure(figsize=(10, 6))
sns.boxplot(hue=df['condition'], y=df['price'], palette='coolwarm')

plt.title('Состояние домов и их стоимость')
plt.xlabel('Состояние домов')
plt.ylabel('Стоимость')
plt.grid(True)
plt.show()
````

Вывод:

*   На коробчатой диаграмме видно, что состояние домов действительно влияет на их стоимость. Дома в отличном состоянии (оценка 5) имеют более высокие медианные цены по сравнению с домами в удовлетворительном состоянии.




````python
plt.figure(figsize=(10, 6))
sns.violinplot(hue=df['bedrooms'], y=df['price'], palette='muted')

plt.title('Количество спален и стоимость')
plt.xlabel('Количество спален')
plt.ylabel('Стоимость')
plt.grid(True)
plt.show()
````

Вывод:

*   График показывает, что стоимость домов увеличивается с увеличением количества спален. Дома с 3-4 спальнями имеют более широкое распределение цен, что предполагает наличие как доступных, так и более дорогих объектов в этом сегменте. Однако для домов с 5 и более спальнями распределение сдвинуто в сторону более высоких цен, что указывает на их принадлежность к премиальному сегменту рынка недвижимости.




````python
plt.figure(figsize=(8, 6))
key_features = ['price', 'sqft_living', 'grade', 'bathrooms', 'bedrooms']
corr_matrix = df[key_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0)

plt.title('Корреляция стоимости с ключевыми характеристиками')
plt.show()
````


Вывод:
*   Обновленная тепловая карта показывает, что наибольшая корреляция с ценой наблюдается у жилой площади (sqft_living) и оценки качества дома (grade). Количество ванных комнат (bathrooms) также имеет значительную корреляцию со стоимостью, хотя она немного слабее.

*Интересно, что количество спален (bedrooms) демонстрирует более низкую корреляцию с ценой, что может означать, что сами по себе спальни не оказывают такого сильного влияния на цену, как общая площадь или качество дома.*
